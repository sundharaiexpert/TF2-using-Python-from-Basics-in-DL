```python
plt.plot(x,y)

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Title of diagram')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
plt.plot(x,y, color = 'red')

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Title of diagram')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
plt.plot(x,y, color='#bc67c9', marker='o', markersize=15, linestyle='--')

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Title of diagram')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```
